#### Load the python libraries

```python
import arviz as az
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
```

#### Load the posterior distributions of all three models

```python
posterior_1 = pd.read_csv("bay_fit_1.csv", index_col=0)
posterior_2 = pd.read_csv("bay_fit_2.csv", index_col=0)
posterior_3 = pd.read_csv("bay_fit_3.csv", index_col=0)
```

```python
coefficient_names1 = posterior_1.columns
#.columns[:11]
#coefficient_names2 = posterior_1.columns[11:]
```

```python
post_summary11 = posterior_1.loc[:,coefficient_names1].T
post_summary11 = post_summary11.set_index(coefficient_names1)

log_dict11 = {}
for i, row in post_summary11.iterrows():
    log_dict11[i] = list(np.array(post_summary11.loc[i,:]))

post_summary21 = posterior_2.loc[:,coefficient_names1].T
post_summary21= post_summary21.set_index(coefficient_names1)

log_dict21 = {}
for i, row in post_summary21.iterrows():
    log_dict21[i] = list(np.array(post_summary21.loc[i,:]))

post_summary31 = posterior_3.loc[:,coefficient_names1].T
post_summary31 = post_summary31.set_index(coefficient_names1)

log_dict31 = {}
for i, row in post_summary31.iterrows():
    log_dict31[i] = list(np.array(post_summary31.loc[i,:]))
```

```python
fig, ax = plt.subplots(figsize=(8, 10))
az.plot_forest(
    log_dict11,
    kind="ridgeplot",
    combined=True,
    var_names = ["~b_Height", "~b_Weight"],
    ridgeplot_truncate=True,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="yellow",
    ridgeplot_alpha=0.5,
    ax=ax
)
az.plot_forest(
    log_dict21,
    kind="ridgeplot",
    combined=True,
    var_names = ["~b_Height", "~b_Weight"],
    ridgeplot_truncate=True,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="red",
    ridgeplot_alpha=0.5,
    ax=ax
)
az.plot_forest(
    log_dict31,
    kind="ridgeplot",
    combined=True,
    var_names = ["~b_Height", "~b_Weight"],
    ridgeplot_truncate=True,
    ridgeplot_quantiles=[0.5],
    markersize=1,
    ridgeplot_overlap=2,
    colors="blue",
    ridgeplot_alpha=0.5,
    ax=ax
)

yellow_patch = mpatches.Patch(color='yellow', label='bay_fit_1', alpha=0.5)
red_patch = mpatches.Patch(color='red', label='bay_fit_2', alpha=0.5)
blue_patch = mpatches.Patch(color='blue', label='bay_fit_3', alpha=0.5)
plt.legend(handles=[yellow_patch, red_patch, blue_patch], fontsize = 10)

ax.axvline(x=0)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.xlabel("Posterior distribution", fontdict={"fontsize": 10})
plt.savefig("images/feature_dist.pdf", dpi = 800,  bbox_inches="tight")
plt.show()
```
