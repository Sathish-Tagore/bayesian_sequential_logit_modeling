```python
import arviz as az
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
model_1 = pd.read_csv("posterior_1.csv")
model_2 = pd.read_csv("posterior_2.csv")
model_3 = pd.read_csv("posterior_3.csv")
model_4 = pd.read_csv("posterior_4.csv")
```

```python
posterior_1 = model_1.drop('b_Weight', axis=1)
posterior_2 = model_2.drop('b_Weight', axis=1)
posterior_3 = model_3.drop('b_Weight', axis=1)
posterior_4 = model_4.drop('b_Weight', axis=1)
Weight_model1 = model_1['b_Weight']
Weight_model2 = model_2['b_Weight']
Weight_model3 = model_3['b_Weight']
Weight_model4 = model_4['b_Weight']
```

```python
coefficient_names1 = posterior_1.columns[:11]
coefficient_names2 = posterior_1.columns[11:]
```

```python
post_summary11 = posterior_1.loc[:,coefficient_names1].T
post_summary11 = post_summary11.set_index(coefficient_names1)

log_dict11 = {}
for i, row in post_summary11.iterrows():
    log_dict11[i] = list(np.array(post_summary11.loc[i,:]))

post_summary21 = posterior_2.loc[:,coefficient_names1].T
post_summary21= post_summary21.set_index(coefficient_names1)

log_dict21 = {}
for i, row in post_summary21.iterrows():
    log_dict21[i] = list(np.array(post_summary21.loc[i,:]))

post_summary31 = posterior_3.loc[:,coefficient_names1].T
post_summary31 = post_summary31.set_index(coefficient_names1)

log_dict31 = {}
for i, row in post_summary31.iterrows():
    log_dict31[i] = list(np.array(post_summary31.loc[i,:]))

post_summary41 = posterior_4.loc[:,coefficient_names1].T
post_summary41 = post_summary41.set_index(coefficient_names1)

log_dict41 = {}
for i, row in post_summary41.iterrows():
    log_dict41[i] = list(np.array(post_summary41.loc[i,:]))
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
az.plot_forest(
    log_dict11,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="green",
    ridgeplot_alpha=0.3,
    ax=ax
)
az.plot_forest(
    log_dict21,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="red",
    ridgeplot_alpha=0.3,
    ax=ax
)
az.plot_forest(
    log_dict31,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="blue",
    ridgeplot_alpha=0.3,
    ax=ax
)

az.plot_forest(
    log_dict41,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="yellow",
    ridgeplot_alpha=0.3,
    ax=ax
)

green_patch = mpatches.Patch(color='green', label='normal(0,100)', alpha=0.3)
red_patch = mpatches.Patch(color='red', label='normal(0,10)', alpha=0.3)
blue_patch = mpatches.Patch(color='blue', label='normal(0,5)', alpha=0.3)
yellow_patch = mpatches.Patch(color="yellow", label="normal(0,1)", alpha=0.3)
plt.legend(handles=[green_patch, red_patch, blue_patch, yellow_patch], fontsize = 6)

ax.grid()
ax.axvline(x=0)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
plt.yticks(fontsize=8)
plt.xticks(fontsize=6)
#plt.savefig("presentations/feature_dist1.png", dpi = 800,  bbox_inches="tight")
plt.show()
```

```python
post_summary12 = posterior_1.loc[:,coefficient_names2].T
post_summary12 = post_summary12.set_index(coefficient_names2)

log_dict12 = {}
for i, row in post_summary12.iterrows():
    log_dict12[i] = list(np.array(post_summary12.loc[i,:]))

post_summary22 = posterior_2.loc[:,coefficient_names2].T
post_summary22= post_summary22.set_index(coefficient_names2)

log_dict22 = {}
for i, row in post_summary22.iterrows():
    log_dict22[i] = list(np.array(post_summary22.loc[i,:]))

post_summary32 = posterior_3.loc[:,coefficient_names2].T
post_summary32 = post_summary32.set_index(coefficient_names2)

log_dict32 = {}
for i, row in post_summary32.iterrows():
    log_dict32[i] = list(np.array(post_summary32.loc[i,:]))

post_summary42 = posterior_4.loc[:,coefficient_names2].T
post_summary42 = post_summary42.set_index(coefficient_names2)

log_dict42 = {}
for i, row in post_summary42.iterrows():
    log_dict42[i] = list(np.array(post_summary42.loc[i,:]))
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
az.plot_forest(
    log_dict12,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="green",
    ridgeplot_alpha=0.3,
    ax=ax
)
az.plot_forest(
    log_dict22,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="red",
    ridgeplot_alpha=0.3,
    ax=ax
)
az.plot_forest(
    log_dict32,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="blue",
    ridgeplot_alpha=0.3,
    ax=ax
)

az.plot_forest(
    log_dict42,
    kind="ridgeplot",
    combined=True,
    ridgeplot_truncate=False,
    ridgeplot_quantiles=[0.5],
    markersize=2,
    ridgeplot_overlap=2,
    colors="yellow",
    ridgeplot_alpha=0.3,
    ax=ax
)

green_patch = mpatches.Patch(color='green', label='normal(0,100)', alpha=0.3)
red_patch = mpatches.Patch(color='red', label='normal(0,10)', alpha=0.3)
blue_patch = mpatches.Patch(color='blue', label='normal(0,5)', alpha=0.3)
yellow_patch = mpatches.Patch(color="yellow", label="normal(0,1)", alpha=0.3)
plt.legend(handles=[green_patch, red_patch, blue_patch, yellow_patch], fontsize = 6)

ax.grid()
ax.axvline(x=0)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
plt.yticks(fontsize=8)
plt.xticks(fontsize=6)
#plt.savefig("presentations/feature_dist1.png", dpi = 800,  bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.kdeplot(Weight_model1, alpha = 0.3, color="green", fill="True", label='normal(0,100)')
sns.kdeplot(Weight_model2, alpha = 0.3, color="red", fill="True", label='normal(0,10)')
sns.kdeplot(Weight_model3, alpha = 0.3, color="blue", fill="True", label='normal(0,5)')
sns.kdeplot(Weight_model4, alpha = 0.3, color="yellow", fill="True", label='normal(0,1)')
green_patch = mpatches.Patch(color='green', label='normal(0,100)', alpha=0.3)
red_patch = mpatches.Patch(color='red', label='normal(0,10)', alpha=0.3)
blue_patch = mpatches.Patch(color='blue', label='normal(0,5)', alpha=0.3)
yellow_patch = mpatches.Patch(color="yellow", label="normal(0,1)", alpha=0.3)
plt.legend(handles=[green_patch, red_patch, blue_patch, yellow_patch], fontsize = 6)
plt.yticks(fontsize=6)
plt.xticks(fontsize=6)
plt.xlabel("b_Weight",fontdict={"fontsize":6})
plt.ylabel("Density",fontdict={"fontsize":6})
plt.show()
```
