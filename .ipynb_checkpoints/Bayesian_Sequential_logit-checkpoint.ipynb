#### Transform the categorical variables into factor variables
* family_history_with_overweight
* FAVC - Frequent consumption of high caloric food
* FCVC - Frequency of consumption of vegetables
* NCP - Number of main meals
* CAEC - Consumption of food between meals
* CH20 - Consumption of water daily
* CALC - Consumption of alcohol

#### The attributes related with the physical condition are:

* SCC - Calories consumption monitoring
* FAF - Physical activity frequency
* TUE - Time using technology devices
* MTRANS - Transportation used
* CALC - Variable "Frequently" is changed to "Always"

```r
data$NObeyesdad <- factor(data$NObeyesdad, levels = c("Insufficient_Weight", 
                                                      "Normal_Weight",
                                                      "Overweight_Level_I",
                                                      "Overweight_Level_II",
                                                      "Obesity_Type_I",
                                                      "Obesity_Type_II",
                                                      "Obesity_Type_III"), ordered = T)
data$Gender <- factor(data$Gender)
data$family_history_with_overweight <- factor(data$family_history_with_overweight)
data$FAVC <- factor(data$FAVC)
data$CAEC <- factor(data$CAEC)
data$SMOKE <- factor(data$SMOKE)
data$SCC <- factor(data$SCC)
data$CALC <- factor(data$CALC)
data$MTRANS <- factor(data$MTRANS)
```

#### Summary of the data

```r
summary(data)
```

```r
# For CALC , Category Always has 1 entry, so changing it into Frequently
levels(data$CALC)[levels(data$CALC) == 'Always'] <- 'Frequently'
```

#### Scale the numerical columns

```r
numeric_columns <- sapply(data, is.numeric)
data[numeric_columns] <- scale(data[numeric_columns])
```

#### Split train and validation set

```r
set.seed(100)
# Generate random samples for training and test sets
samplesize <- 0.70 * nrow(data)
index <- sample(seq_len(nrow(data)), size = samplesize)

# Creating training and test sets
datatrain <- data[index, ]
datatest <- data[-index, ]
```

#### Bayesian Fit - 1 with weakly informative prior

```r
# First Bayesian fit
prior_1 <- prior(normal(0, 5), class = "b") + prior(normal(0, 5), class = "Intercept")
bay_fit_1 <- brm(NObeyesdad ~ ., data = datatrain, family = sratio(), prior = prior_1, seed = 100)
```

```r
#Posterior estimates
summary(bay_fit_1)
```

```r
#Make predictions on test set
predictions1 <- predict(bay_fit_1, newdata = datatest, summary = FALSE)
conf_matrix1 <- table(datatest$NObeyesdad, predictions1[1,])
print(conf_matrix1)
```

```r
#Calculate the accuracy of the fit
accuracy1 <- sum(diag(conf_matrix1)) / sum(conf_matrix1)
cat("Accuracy:", accuracy1, "\n")
```

#### Bayesian Fit - 2 with informative default cauchy prior

```r
# Second Bayesian fit
prior_2 <- prior(cauchy(0, 2.5), class = "b") + prior(normal(0, 5), class = "Intercept")
bay_fit_2 <- brm(NObeyesdad ~ ., data = datatrain, family = sratio(), prior = prior_2, seed = 100)
```

```r
#Posterior estimates
summary(bay_fit_2)
```

```r
#Make predictions on test set
predictions2 <- predict(bay_fit_2, newdata = datatest, summary = FALSE)
conf_matrix2 <- table(datatest$NObeyesdad, predictions2[1,])
print(conf_matrix2)
```

```r
#Calculate the accuracy of the fit
accuracy2 <- sum(diag(conf_matrix2)) / sum(conf_matrix2)
cat("Accuracy:", accuracy2, "\n")
```

#### Bayesian Fit - 3 with informative cauchy prior with small scale

```r
# Third Bayesian fit
prior_3 <- prior(cauchy(0, 0.1), class = "b") + prior(normal(0, 5), class = "Intercept")
bay_fit_3 <- brm(NObeyesdad ~ ., data = datatrain, family = sratio(), prior = prior_3, seed = 100)
```

```r
#Posterior estimates
summary(bay_fit_3)
```

```r
#Make predictions on test set
predictions3 <- predict(bay_fit_3, newdata = datatest, summary = FALSE)
conf_matrix3 <- table(datatest$NObeyesdad, predictions3[1,])
print(conf_matrix3)
```

```r
#Calculate the accuracy of the fit
accuracy3 <- sum(diag(conf_matrix3)) / sum(conf_matrix3)
cat("Accuracy:", accuracy3, "\n")
```

#### Leave one out cross validation

```r
loo_ma1 <- loo(bay_fit_1)
loo_ma2 <- loo(bay_fit_2)
loo_ma3 <- loo(bay_fit_3)
loo_compare_result <- loo_compare(loo_ma1, loo_ma2, loo_ma3)
print(loo_compare_result, simplify = F, digits = 3)
```

```r
est_par = c('b_GenderMale',
            'b_Age',
            'b_Height',
            'b_Weight',
            'b_family_history_with_overweightyes',
            'b_FAVCyes','b_FCVC','b_NCP','b_CAECFrequently',
            'b_CAECSometimes',
            'b_CAECno',
            'b_SMOKEyes',
            'b_CH2O',
            'b_SCCyes',
            'b_FAF',
            'b_TUE',
            'b_CALCSometimes',
            'b_CALCno',
            'b_MTRANSBike',
            'b_MTRANSMotorbike',
            'b_MTRANSPublic_Transportation',
            'b_MTRANSWalking')

# Extract posteriors for each model
posterior_1 <- as.data.frame(as.matrix(bay_fit_1))[est_par]
posterior_2 <- as.data.frame(as.matrix(bay_fit_2))[est_par]
posterior_3 <- as.data.frame(as.matrix(bay_fit_3))[est_par]

write.csv(posterior_1, file = "bay_fit_1.csv")
write.csv(posterior_2, file = "bay_fit_2.csv")
write.csv(posterior_3, file = "bay_fit_3.csv")
```
